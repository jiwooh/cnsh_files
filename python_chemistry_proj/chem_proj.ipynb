```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1. csv 불러오기

[csv 파일 링크](https://gist.github.com/robertwb/22aa4dbfb6bcecd94f2176caa912b952)

```python
pt=pd.read_csv('https://gist.githubusercontent.com/robertwb/22aa4dbfb6bcecd94f2176caa912b952/raw/2dfcd7ebb235bdd9fb13b955128389c0b9efe394/Periodic%2520Table%2520of%2520Elements.csv').fillna(np.NaN)
pt
```

```python
Symbol=pt['Symbol'].tolist()
Symbol.insert(0,'')
```

# 2. 원자량

```python
AM=pt['AtomicMass']
AM.loc[-1]=np.NaN
AM.index=AM.index+1
AM.sort_index(inplace=True)
AM
```

```python
plt.figure(figsize=(40,8))
plt.scatter(x=Symbol, y=AM, color='g')
plt.xlim([0,119])
plt.ylim([0,301])
plt.xticks(range(1,119,1))
plt.yticks(range(0,301,20))
plt.xlabel('Element Number', fontsize=15)
plt.ylabel('Atomic Mass', fontsize=15, fontweight='bold')
plt.grid(True)
plt.show()
```

```python
NoN=pt['NumberofNeutrons']
NoN.loc[-1]=np.NaN
NoN.index=NoN.index+1
NoN.sort_index(inplace=True)
NoN
```
