```python
from itertools import permutations as nPr
# from itertools import *
```

# 1. 순서를 생각한 순열

```python
A=['민수', '가원', '한승', '민석', '강준', '동린']
X=list(nPr(A))
res=[]
for i in X:
    if i.index('한승') > i.index('민석') > i.index('동린'):
        res.append(i)
print('* There are {} cases.'.format(len(res))) # 6!/3!=120
print('* The cases are', res)
```
