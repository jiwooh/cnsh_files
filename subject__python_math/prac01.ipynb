```python
from itertools import permutations as nPr
from itertools import product as nÎ r
from itertools import combinations as nCr
from itertools import combinations_with_replacement as nHr
from fractions import Fraction
```

```python
A=[1,1,2,3,4]
T=set(list(nPr(A,4)))
P=[i for i in T if i[0]<=i[1]<=i[2]<=i[3]]
print(Fraction(len(P),len(T)))
```

```python
A=range(2,15)
B=list(nÎ r(A,repeat=4))
C=[i for i in B if all((sum(i)==20, i[1]%i[0]==0, i[2]%i[0]==0, i[3]%i[0]==0))]
print(len(C))
```

```python
A=[1,1,1,0,0,0,0]
B=list(nCr(A,3))
P=[i for i in B if i.count(1)*2>=i.count(0)]
Q=[i for i in B if i.count(1)==2]
print(Fraction(len(Q),len(P)))
```
