```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
weight = 22, 24, 26, 30, 32, 40, 35, 45, 20, 29, 34, 36, 36, 38, 39, 48, 43, 37, 33, 31, 29, 39, 26, 29
bum=np.arange(20,50,5)
wei,bum=np.histogram(weight,bum)
plt.hist(weight, bum, rwidth=0.7, color='#d8a1fb', alpha=0.5, label='weight')
plt.title('2반 몸무게', fontsize=15, fontweight='bold')
plt.xticks(fontsize=13, fontweight='light')
plt.ylabel('명수', fontsize=12)
plt.grid()
plt.legend()
plt.show()
```

```python
b=np.arange(1,20,3)
c=np.arange(3,50,7)
print(b, c)
plt.scatter(b, c)
plt.title('\n산포도 test', fontsize=20)
plt.show()
```

```python
st=[]
a=0
for i in wei:
    a+=i
    st.append(a)
print('누적도수:', st)
```

```python
d=bum[:-1]+2.5
plt.scatter(d,st)
plt.plot(d, st)
plt.grid()
plt.yticks(range(0,25,2), fontsize=12)
plt.show()
```

```python
P=np.random.randint(1,101,10)
print(P, '\n')

avr=sum(P)/len(P)
print('평균, 식으로 구하기:', avr)
print('평균, numpy:', np.average(P))

var=sum((P-avr)**2)/len(P)
print('분산, 식으로 구하기:', var)
print('분산, 공식쓰기:', sum(P**2)/len(P**2)-(sum(P)/len(P))**2)
print('분산, numpy:', np.var(P))

std=np.sqrt(var)
print('표준편차:', std)
print('표준편차:', np.std(P))
```
